```python
#
years = list(range(2024,2022, -1))
```

```python
years
```

```python
import time

#Iterate over each year
for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text, 'html.parser')
    league_table = soup.select('table.stats_table')[0]

    links = [l.get("href") for l in league_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]# Filter links to only include those related to team squads
    team_urls = [f"https://fbref.com{l}" for l in links]
    
    # Get the URL for the previous season standings
    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"
    
    # Iterate over each team's URL
    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")
        data = requests.get(team_url)
        matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
        soup = BeautifulSoup(data.text, 'html.parser')
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l] #Filter links to only include those related to shooting statistics
        data = requests.get(f"https://fbref.com{links[0]}")
        shooting = pd.read_html(data.text, match="Shooting")[0] # Extract shooting statistics
        shooting.columns = shooting.columns.droplevel()

        #sometimes data is missing so we will use a try block to ignore these scenarios
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError:
            continue
            
        # Filter data to only include Premier League matches
        team_data = team_data[team_data["Comp"] == "Premier League"]
        
        # Add season and team name columns to the data
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        
        # Pause for 15 seconds to avoid overwhelming the server with requests
        time.sleep(15)
```
